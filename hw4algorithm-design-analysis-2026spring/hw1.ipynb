## 问题 1

根据下列递归公式, 计算下列 $T(n)$ 对应的的渐近上界。要求所求的边界尽可能的紧（tight）, 请写明步骤。

1. $T(1)=1; T(n)=T(n/4)+1$ for $n>1$
2. $T(1)=1;T(n)=3T(n/3)+n^2$ for $n>1$
3. $T(1)=1;T(n)=T(2n/3)+1$ for $n>1$
4. $T(1)=1;T(n)=5T(n/4)+n$ for $n>1$
5. $T(n)=1 \ for\ n \le 2 ; T(n)=T(\sqrt{n})+1 \ for \ n>2$

解答：

下面忽略取整带来的常数影响，只讨论渐近紧界。

1. $T(n)=T(n/4)+1$

   每递归一次，问题规模变为原来的 $1/4$。设递归深度为 $k$，当 $n/4^k=1$ 时停止，所以

   $$k=\log_4 n.$$

   每层只做 $O(1)$ 工作，因此

   $$T(n)=\Theta(\log_4 n)=\Theta(\log n).$$

2. $T(n)=3T(n/3)+n^2$

   使用主定理：$a=3,b=3$，所以

   $$n^{\log_b a}=n^{\log_3 3}=n.$$

   非递归部分 $f(n)=n^2$，比 $n$ 多一个多项式因子，并且满足 regularity 条件：

   $$3f(n/3)=3(n/3)^2=n^2/3 \le c n^2,$$

   其中 $c<1$。因此由主定理第 3 种情况可得

   $$T(n)=\Theta(n^2).$$

3. $T(n)=T(2n/3)+1$

   每次递归后规模变为 $(2/3)n$。设递归深度为 $k$，当 $(2/3)^k n=1$ 时停止，即

   $$k=\log_{3/2} n.$$

   每层代价为 $O(1)$，所以

   $$T(n)=\Theta(\log_{3/2} n)=\Theta(\log n).$$

4. $T(n)=5T(n/4)+n$

   使用主定理：$a=5,b=4$，所以

   $$n^{\log_b a}=n^{\log_4 5}.$$

   因为 $f(n)=n=O(n^{\log_4 5-\epsilon})$，其中可以取任意 $0<\epsilon<\log_4 5-1$，所以由主定理第 1 种情况可得

   $$T(n)=\Theta(n^{\log_4 5}).$$

5. $T(n)=T(\sqrt n)+1$

   令 $n=2^m$，并定义 $S(m)=T(2^m)$。因为 $\sqrt n=2^{m/2}$，所以递归式变为

   $$S(m)=S(m/2)+1.$$

   由第 1 类递归可知

   $$S(m)=\Theta(\log m).$$

   又因为 $m=\log_2 n$，所以

   $$T(n)=\Theta(\log\log n).$$

## 问题 3

给定一个包含n个元素的数组 `profits` , 它的第 `i` 个元素 `profits[i]` 表示一支股票第 `i` 天的**收益**（正数表示涨, 负数表示跌）。你只能选择 **某一天** 买入这只股票, 并选择在 **未来的某一个不同的日子** 卖出该股票。

1. 设计一个算法来计算你所能获取的最大利润和对应买入和卖出的日期。请分析算法方案, 计算其时间复杂度, 并且使用python编程实现该算法。

2. \* 设计一个时间复杂度为 $O(n)$的算法实现该算法

e.g. :
---
profits=[3,2,1,-7,5,2,-1,3,-1], 第5天买入, 第8天卖出, 收益最大：9

idea:(格式由AI辅助生成)

最简单的方法可以直接暴力枚举所有买入、卖出日期的方法。因为 `profits[i]` 表示第 `i` 天的收益，选择第 `i` 天买入、第 `j` 天卖出时，总收益就是连续区间 `profits[i..j]` 的和。枚举每个买入日，再向后枚举卖出日并维护当前区间和，即可找到最大收益和对应日期。

算法步骤：

1. 对每个可能的买入日 `i`，令当前收益为 `profits[i]`。
2. 枚举未来卖出日 `j>i`，不断把 `profits[j]` 加入当前收益。
3. 若当前收益更大，则更新最大收益、买入日和卖出日。

时间复杂度：两层循环枚举所有区间，复杂度为 $O(n^2)$。

空间复杂度：只使用n个变量，复杂度为 $O(n)$。


````python
def max_profit_bruteforce(profits):
    n = len(profits)
    if n < 2:
        print("至少需要两天的数据来计算利润。")
        return 0

    best_profit = float("-inf")
    best_buy = best_sell = None

    for i in range(n - 1):
        current = profits[i]
        for j in range(i + 1, n):
            current += profits[j]
            if current > best_profit:
                best_profit = current
                best_buy = i + 1
                best_sell = j + 1

    return best_profit, best_buy, best_sell


profits = [3, 2, 1, -7, 5, 2, -1, 3, -1]
profit, buy, sell = max_profit_bruteforce(profits)
print(f"最大收益: {profit}, 买入日期: 第{buy}天, 卖出日期: 第{sell}天")
````

给定一个包含不同整数元素的数组 $ A[1..n]$ ,并且满足条件：$A[1]>A[2]$ 并且 $A[n-1]<A[n]$; 规定：如果一个元素比它两边的邻居元素都小, 即：$A[x]<A[x-1], A[x]<A[x+1]$ , 称这个元素A[x]为“局部最小”。通过遍历一次数组, 我们可以很容易在 $O(n)$的时间复杂度下找到一个局部最小值, 


1. 分析该问题, 设计一个算法在$O(logn)$的时间复杂度下找到一个局部最小(返回数值), 要求：分析算法设计思路, 并且使用python编程实现
2. \* 设计算法找出所有局部最小值, 分析算法设计思路, 并使用python编程实现

e.g.:
---
A=[9, 3, 7, 2, 1, 4, 5 ] 时,  局部最小元素为 3, 1

idea:(格式由AI辅助生成)

数组元素互不相同，并且满足 $A[1]>A[2]$、$A[n-1]<A[n]$，因此数组内部一定存在局部最小。可以用二分法寻找一个局部最小：

1. 只在内部下标范围 `[1, n-2]` 中搜索。
2. 若 `A[mid]` 同时小于左右邻居，则它就是局部最小。
3. 若 `A[mid] > A[mid-1]`，说明从 `mid` 向左存在下降方向，左侧一定存在局部最小，因此搜索左半边。
4. 否则 `A[mid] > A[mid+1]`，右侧存在下降方向，搜索右半边。

每次搜索区间减半，所以时间复杂度为 $O(\log n)$；只使用n个变量，空间复杂度为 $O(n)$。


````python
def find_local_minimum(A):
    if len(A) < 3:
        raise ValueError("数组长度必须至少为3")
    if not (A[0] > A[1] and A[-2] < A[-1]):
        raise ValueError("数组必须满足 A[0] > A[1] 和 A[-2] < A[-1]")

    left, right = 1, len(A) - 2
    while left <= right:
        mid = (left + right) // 2
        if A[mid] < A[mid - 1] and A[mid] < A[mid + 1]:
            return A[mid]
        if A[mid] > A[mid - 1]:
            right = mid - 1
        else:
            left = mid + 1

    raise RuntimeError("未找到局部最小值")


A = [9, 3, 7, 2, 1, 4, 5]
print("找到的一个局部最小值:", find_local_minimum(A))
````

给定包含n个不同数字的一组数, 寻找一种基于比较的算法在这组数中找到k个最小的数字, 并按顺序输出它们。

1. 将n个数先进行排序, 然后按顺序输出最小的k个数。要求：选择合适的排序算法实现上述操作, 计算算法时间复杂度, 并使用python编程实现。
2. 建立一个包含这n个数的堆（heap）, 并且调用 k 次Extract-min 按顺序输出最小的k个数。使用往空堆中不断插入元素的方法建立堆, 分析这种方法建堆的时间复杂度, 并使用python编程实现
3. \* 假设数组中包含的数据总数目超过了计算机的存储能力, 请设计一个算法, 找到这堆数据的前k小的数值, 计算时间复杂度, 并使用python实现该算法, 假设计算机一定能存储k个数据。

e.g.：
---
数组arr=[5,4,3,2,6,1,88,33,22,107] 的前3个最小数据为：1, 2, 3

idea:(格式由AI辅助生成)

1. 先排序再输出前 $k$ 个数：可以使用归并排序。归并排序时间复杂度为 $O(n\log n)$，排序后取前 $k$ 个元素需要 $O(k)$，因此总时间复杂度为 $O(n\log n+k)$，通常记为 $O(n\log n)$。归并排序需要 $O(n)$ 额外空间。

2. 使用堆：按照题目要求，先从空堆开始，把 $n$ 个元素逐个插入最小堆。每次插入的代价是 $O(\log n)$，所以这种建堆方式的时间复杂度是 $O(n\log n)$。之后执行 $k$ 次 `Extract-min`，每次复杂度 $O(\log n)$，所以输出前 $k$ 小元素的代价是 $O(k\log n)$。总时间复杂度为 $O((n+k)\log n)$，空间复杂度为 $O(n)$。



````python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr[:]

    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])

    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result.extend(left[i:])
    result.extend(right[j:])
    return result


def k_smallest_by_sort(arr, k):
    if k < 0 or k > len(arr):
        raise ValueError("k必须在0和len(arr)之间")
    return merge_sort(arr)[:k]


class MinHeap:
    def __init__(self):
        self.data = []

    def insert(self, value):
        self.data.append(value)
        self._sift_up(len(self.data) - 1)

    def extract_min(self):
        if not self.data:
            raise IndexError("从空堆中提取最小值")
        minimum = self.data[0]
        last = self.data.pop()
        if self.data:
            self.data[0] = last
            self._sift_down(0)
        return minimum

    def _sift_up(self, idx):
        while idx > 0:
            parent = (idx - 1) // 2
            if self.data[parent] <= self.data[idx]:
                break
            self.data[parent], self.data[idx] = self.data[idx], self.data[parent]
            idx = parent

    def _sift_down(self, idx):
        n = len(self.data)
        while True:
            left = 2 * idx + 1
            right = 2 * idx + 2
            smallest = idx
            if left < n and self.data[left] < self.data[smallest]:
                smallest = left
            if right < n and self.data[right] < self.data[smallest]:
                smallest = right
            if smallest == idx:
                break
            self.data[idx], self.data[smallest] = self.data[smallest], self.data[idx]
            idx = smallest


def k_smallest_by_heap(arr, k):
    if k < 0 or k > len(arr):
        raise ValueError("k必须在0和len(arr)之间")

    heap = MinHeap()
    for value in arr:
        heap.insert(value)
    return [heap.extract_min() for _ in range(k)]


arr = [5, 4, 3, 2, 6, 1, 88, 33, 22, 107]
k = 3
print("排序方法:", k_smallest_by_sort(arr, k))
print("堆方法:", k_smallest_by_heap(arr, k))
````

## 问题 6

> 星球碰撞问题：现有n个星球，在同一条直线上运行，如数组A所示，元素的绝对值表示星球的质量，负数表示星球自右向左运动，正数表示星球自左向右运动，当两个星球相撞的时候，质量小的会消失，大的保持不变，**质量相同的两个星球碰撞后自右向左运动的星球消失，自左向右的星球保持不变**，假设所有星球的速度大小相同。
>
> $ A=[23,-8, 9, -3, -7, 9, -23, 22] $

请设计一个算法模拟星球的运行情况，输出最终的星球存续情况（输出一个数组），分析算法设计思路，计算时间复杂度，并基于python编程实现。

e.g.
---
1.  输入： A=[-3,-6,2,8, 5,-8,9,-2,1]， 输出：[-3, -6, 2, 8, 9, 1]

2. 输入：A=[23,-8, 9, -3, -7, 9, -23, 22], 输出：[23, 22]

idea:(格式由AI辅助生成)

使用栈模拟从左到右运动的过程。只有一种情况会发生碰撞：栈顶星球向右运动（正数），当前星球向左运动（负数）。其他方向组合不会迎面相撞。

处理当前星球 `x` 时：

1. 如果栈顶为正且 `x` 为负，则比较二者质量。
2. 若栈顶质量小于 `x`，栈顶星球消失，继续让 `x` 和新的栈顶比较。
3. 若栈顶质量大于 `x`，当前星球消失。
4. 若质量相等，按题目规则，自右向左运动的星球消失，也就是当前负数星球消失，栈顶正数保留。
5. 如果当前星球最终没有消失，则入栈。

每个星球最多入栈一次、出栈一次，因此时间复杂度为 $O(n)$，空间复杂度为 $O(n)$。

````python
def surviving_planets(A):
    stack = []

    for x in A:
        alive = True
        while alive and stack and stack[-1] > 0 and x < 0:
            top_mass = abs(stack[-1])
            current_mass = abs(x)
            if top_mass < current_mass:
                stack.pop()
            else:
                alive = False
        if alive:
            stack.append(x)

    return stack


case1 = [-3, -6, 2, 8, 5, -8, 9, -2, 1]
case2 = [23, -8, 9, -3, -7, 9, -23, 22]
print(surviving_planets(case1))
print(surviving_planets(case2))
````

## 问题 7

> 0-1背包问题：现在有4块大理石，每个大理石对应的重量和价值使用一个元组表示, 即（重量，价值），4块大理石的重量和价值为：[(5,10), (4,40), (6,30), (3, 50)], 假设你有一辆最大承重为9的小推车，请问使用这个小推车装大理石的最大价值为多少

请设计一个算法解决该问题，分析算法设计思路，计算时间复杂度，并基于python编程实现

idea:(格式与其中公式由AI辅助生成)

这是标准 0-1 背包问题。每块大理石只能选或不选一次，因此可以使用动态规划。

定义 `dp[i][w]` 表示只考虑前 `i` 块大理石、容量为 `w` 时能够获得的最大价值。对于第 `i` 块大理石，设重量为 `weight`、价值为 `value`：

1. 如果 `weight > w`，不能选它，`dp[i][w] = dp[i-1][w]`。
2. 否则可以在“不选”和“选”之间取最大值：

   $$dp[i][w]=\max(dp[i-1][w], dp[i-1][w-weight]+value).$$

最终答案为 `dp[n][capacity]`。时间复杂度为 $O(nW)$，空间复杂度为 $O(nW)$，其中 $W$ 是最大承重。

````python
def knapsack_01(items, capacity):
    n = len(items)
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]

    for i in range(1, n + 1):
        weight, value = items[i - 1]
        for w in range(capacity + 1):
            dp[i][w] = dp[i - 1][w]
            if weight <= w:
                dp[i][w] = max(dp[i][w], dp[i - 1][w - weight] + value)

    chosen = []
    w = capacity
    for i in range(n, 0, -1):
        if dp[i][w] != dp[i - 1][w]:
            chosen.append(i - 1)
            w -= items[i - 1][0]
    chosen.reverse()

    return dp[n][capacity], chosen


items = [(5, 10), (4, 40), (6, 30), (3, 50)]
capacity = 9
max_value, chosen_indices = knapsack_01(items, capacity)
print("最大价值:", max_value)
print("选择的大理石:", [items[i] for i in chosen_indices])
````

idea:(格式由AI辅助生成)

二叉搜索树的中序遍历结果应该是严格递增序列。如果恰好有两个节点的值被交换，那么中序序列中会出现逆序位置。

算法步骤：

1. 对树做中序遍历，维护上一个访问到的节点 `prev`。
2. 如果发现 `prev.val > current.val`，说明出现逆序。
3. 第一次出现逆序时，错误节点之一是 `prev`，另一个候选是 `current`。
4. 如果再次出现逆序，则第二个错误节点更新为 `current`。
5. 遍历结束后交换这两个节点的值，树结构保持不变。

无论两个错误节点是否相邻，该方法都能恢复。时间复杂度为 $O(n)$，递归栈空间复杂度为 $O(h)$，其中 $h$ 是树高。

````python
from collections import deque


class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def build_tree(values):
    if not values or values[0] is None:
        return None

    root = TreeNode(values[0])
    queue = deque([root])
    idx = 1
    while queue and idx < len(values):
        node = queue.popleft()
        if idx < len(values) and values[idx] is not None:
            node.left = TreeNode(values[idx])
            queue.append(node.left)
        idx += 1
        if idx < len(values) and values[idx] is not None:
            node.right = TreeNode(values[idx])
            queue.append(node.right)
        idx += 1
    return root


def serialize_tree(root):
    if root is None:
        return []

    result = []
    queue = deque([root])
    while queue:
        node = queue.popleft()
        if node is None:
            result.append(None)
        else:
            result.append(node.val)
            queue.append(node.left)
            queue.append(node.right)

    while result and result[-1] is None:
        result.pop()
    return result


def recover_bst(root):
    first = second = prev = None

    def inorder(node):
        nonlocal first, second, prev
        if node is None:
            return
        inorder(node.left)
        if prev is not None and prev.val > node.val:
            if first is None:
                first = prev
            second = node
        prev = node
        inorder(node.right)

    inorder(root)
    if first is not None and second is not None:
        first.val, second.val = second.val, first.val
    return root


root1 = build_tree([1, 3, None, None, 2])
recover_bst(root1)
print(serialize_tree(root1))

root2 = build_tree([3, 1, 4, None, None, 2])
recover_bst(root2)
print(serialize_tree(root2))
````

可以用深度优先搜索遍历从根节点到叶节点的所有路径。遍历到某个节点时，维护从根到当前节点形成的数字 `current`。如果当前节点值为 `x`，则新的路径数字为：

$$current \times 10 + x.$$

当访问到叶节点时，说明一条完整路径已经形成，把该数字加入答案。递归遍历所有节点即可。

每个节点只访问一次，所以时间复杂度为 $O(n)$；递归栈空间复杂度为 $O(h)$，其中 $h$ 是树高。

````python
from collections import deque


class SumTreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def build_sum_tree(values):
    if not values or values[0] is None:
        return None

    root = SumTreeNode(values[0])
    queue = deque([root])
    idx = 1
    while queue and idx < len(values):
        node = queue.popleft()
        if idx < len(values) and values[idx] is not None:
            node.left = SumTreeNode(values[idx])
            queue.append(node.left)
        idx += 1
        if idx < len(values) and values[idx] is not None:
            node.right = SumTreeNode(values[idx])
            queue.append(node.right)
        idx += 1
    return root


def sum_root_to_leaf_numbers(root):
    def dfs(node, current):
        if node is None:
            return 0

        current = current * 10 + node.val
        if node.left is None and node.right is None:
            return current
        return dfs(node.left, current) + dfs(node.right, current)

    return dfs(root, 0)


root1 = build_sum_tree([1, 2, 3])
root2 = build_sum_tree([4, 9, 0, 5, 1])
print(sum_root_to_leaf_numbers(root1))
print(sum_root_to_leaf_numbers(root2))
````
